# Monitoramento de Preços da Gasolina (R$/L)

precos = [5.29, 5.75, 6.10, 5.95, 5.40, 6.25, 5.80]

**Regras de Classificação:**

*  < 5.50 → "Baixo"
*  5.50 – 6.00 → "Moderado"
*  (>6.00 → "Alto"

**Tarefas:**

1.	Filtrar os dias com preço alto
2.	Calcular:
o	Preço total acumulado na semana
o	Preço médio semanal
3.	Exibir alerta se houver mais de 2 dias com preço alto


```python
preços = [5.29, 5.75, 6.10, 5.95, 5.40, 6.25, 5.80]
preço_total = 0
quantidade_preços_altos = 0
for p in preços:
  preço_total += p
  if(p < 5.5):
    print(f"preço {p}: Baixo")
  elif(5.5 <= p <= 6.0):
    print(f"preço {p}: Moderado")
  else:
    print(f"preço {p}: Alto")
    quantidade_preços_altos = quantidade_preços_altos + 1

print(f"Preço total da semana: {preço_total}")
quantidade = len(preços)
média = preço_total / quantidade
print(f"Preço médio da semana: {média:.2f}")

if (quantidade_preços_altos > 2):
  print("Alerta! Muitos dias com preço alto!")
else:
  print("Semana normal.")
```
